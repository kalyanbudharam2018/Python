```python
import pandas as pd
#  One-dimensional labeled array 
data=[1,2,3,4]
print(pd.Series(data))
a=pd.Series(data)
print(type(a))

# How to change the index name? 
print('\n', pd.Series(data, index=('a',' b',' c',' d')))
```

```python
# Create a DataFrame from a List 
d=[1,2,3,4,5]
df=pd.DataFrame(d)
print(df)
print(type(df))
```

```python
#  Create a DataFrame from a Dictionary 
d={'f':['apple', 'mago','banana'], 'g':['one','two', 'three']}
df=pd.DataFrame(d, index=('a','b','c'))
(df)
```

```python
#  Create a DataFrame from a Series 
data = pd.Series([6,12], index=['a','b'])
df = pd.DataFrame([data])
df
```

```python
#  Create a DataFrame from a numpy ND array 
import numpy as np
data= np.array([['a','b'], [6,12]])
df = pd.DataFrame({'A':data[:,0],'B':data[:,1]}) #or df = pd.DataFrame(data, index=('A','B'))
df
```

```python
# DataFrame for Pandas Merge 
player = ['Player1', 'Player2', 'Player3']
points = [8, 9, 5]
title = ['Game1', 'Game2', 'Game3']
df1 = pd.DataFrame({'Player': player,'Points': points,'Title': title}) 
# or df1 = df1[['Player', 'Points', 'Title']]
print(type(df1))
df1
```

```python
player = ['Player1','Player5','Player6']
power = ['Punch','Kick', 'Elbow']
title = ['Game1','Game5','Game6']
df2 = pd.DataFrame({'Player': player, 'Power': power,'Title': title})
df2
     
   
```

```python
# Inner Merge 
df1.merge(df2, on='Title', how='inner') 
```

```python
# Left Merge 
df1.merge(df2, on='Title', how='left')
```

```python
# Right Merge 
df1.merge(df2, on='Title', how='right')
```

```python
# Outer merge 
df1.merge(df2, on='Title', how='outer')
```

```python
# DataFrame for Pandas Join 
# DF1
player = ['Player1', 'Player2', 'Player3']
points = [8, 9, 5]
title = ['Game1', 'Game2', 'Game3']
df1 = pd.DataFrame({'Player': player,'Points': points,'Title': title})
df1.set_index('Player')
df1
```

```python
# DataFrame-2
player = ['Player1','Player5','Player6']
power = ['Punch','Kick', 'Elbow']
title = ['Game1','Game5','Game6']
df2 = pd.DataFrame({'Player': player, 'Power': power,'Title': title})
df2.set_index('Player')
df2
```

```python
# Inner Join
df1.join(df2, how='inner', lsuffix='_x', rsuffix='_y')
```

```python
# Left Join
df1.join(df2, how='left', lsuffix='_x', rsuffix='_y')
```

```python
# Right Join 
df1.join(df2, how='right', lsuffix='_x', rsuffix='_y') 
```

```python
# Outer join 
df1.join(df2, how='outer', lsuffix='_x', rsuffix='_y') 
```

```python
# Concatenate
pd.concat([df1,df2]) 
```
