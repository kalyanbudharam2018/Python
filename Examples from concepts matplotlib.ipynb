```python
# Demonstration: Line Plot 
import numpy as np
import matplotlib.pyplot as plt

# preparing data
a= np.linspace(0,10,20)
b= np.linspace(10,0,20)
# adding figure
fig=plt.figure(figsize=(6,3))
# adding axes
ax1=plt.subplot()
# simple line plot of both a and b
ax1.plot(a)
ax1.plot(b)
# show the plot
plt.show()
```

```python
# Demonstration: Customized Line Plot 
# preparing data
a= np.linspace(0,10,20)
b= np.linspace(10,0,20)
# adding figure
fig=plt.figure(figsize=(6,3))
# adding axes
ax1=plt.subplot()
ax1.plot(a, linewidth=0.5, linestyle='--', color='b',alpha=0.3, marker='^')
ax1.plot(b, linewidth=2.0, linestyle='--', color='r',alpha=1, marker='*')
# x , y labels and title
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line plot')
# legend
plt.legend(['a','b'],loc='best')
# grid
plt.grid(True)
# save the plot
plt.savefig('Lineplot.png')
plt.show()
```

```python
# Demonstration: Sub-plotting 
# preparing data
a= np.linspace(0,10,20)
b= np.linspace(10,0,20)
# adding figure
fig=plt.figure(figsize=(6,3))
# adding sub plots
ax1=plt.subplot(211)  # 2 rows 1 column 1st position
ax2=plt.subplot(212)  # 2 rows 1 column 2nd position

ax1.plot(a, linewidth=0.5, linestyle='--', color='b',alpha=0.3, marker='^')
ax2.plot(b, linewidth=2.0, linestyle='--', color='r',alpha=1, marker='*')

# titles
ax1.set(title='Line plot')
ax2.set(title='LP')

# adding space between subplots
plt.subplots_adjust(left= None, bottom=None, right= None, top= None, wspace=None, hspace=0.5)
plt.show()
```

```python
# Demonstration: Bar Plot 
data={'apple':10, 'orange':15, 'lemons':5}
names=list(data.keys())
values=list(data.values())
#add figure
fig=plt.figure(figsize=(6,3))
# plot
plt.bar(names, values, align='center', color='orange')
# title and labels
plt.title('Bar plot')
plt.xlabel('Names')
plt.ylabel('Values')
# grid
plt.grid()
# save
plt.savefig('Barplot.png')
plt.show()
```

```python
# Demonstration: Horizontal Bar Plot 
plt.barh(names, values, align='center', color='orange')
```

```python
# Demonstration: Scatter Plot 
a=[10,20,30,40,50,60,50,10]
b=[5,4,2,4,6,2,4,3]
x=[1,2,3,4,5,6,8,2]

plt.scatter(a,b,c='g',s=100, edgecolors='y', marker='o',alpha=0.5)
plt.scatter(a,x,c='r',s=500, edgecolors='b', marker='4',alpha=1)

plt.title('Scatter plot')
plt.grid()
plt.show()
```

```python
# Demonstration: Histogram similar for box and violin plots
a=np.linspace(0,100,25, dtype=int)
at=list(a)
plt.hist(at, bins=12)
plt.xlabel('numbers')
plt.ylabel('out')

plt.show()
```
